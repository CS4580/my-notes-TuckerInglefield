#### Asses Seasonality 
When meeting with the emplyees of the Avila Dobe, you hear that one major point is that the number of the visitors varies greatly with the season

Part A:
- Create a line plot that shows how the number of visitors to the Avila Adobe has evolved over time. 
- replace `inf` with `nan` and then fill them in with `0`

```python
import numpy as np 
# Replace infinite values with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)
museum_df = museum_df.fillna(0)
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=museum_df['Avila Adobe'])
plt.title("Monthly Visitors to Avila Adobe")
plt.xlabel("Date")
plt.show()
```
